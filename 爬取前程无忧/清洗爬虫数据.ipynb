```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('51job.csv',encoding='gbk')
df = df.drop_duplicates()
df = df[df.职位.str.contains(r'.*?数据.*?|.*?分析.*?')]#提取包含数据或者分析的岗位
```

```python
def remove_str(df):
    for i in range(0, df.shape[0]):
        for j in range(0, df.shape[1]):
            df.iloc[i][j] = df.iloc[i][j][2:-2]#直接选取字符串中的第一个到最后一个字符，简单方便
            #df.iloc[i][j] = df.iloc[i][j].strip(('['']'))只能去掉外面的中括号
    return df#已经去掉了外面的方框，之后看看怎么把引号去掉
```

```python
df = remove_str(df)
df.head()
df.to_excel('finished.xlsx')
```

```python
df2 = df.薪资.str.extractall('(\d*\.?\d)([-](\d?\.?\d))')
df2.to_excel('df2.xlsx')
```

```python
df3 = pd.read_excel('finished.xlsx',encoding='gbk')
df4 = pd.read_excel('df2.xlsx')
df4.head()
data = pd.concat([df3,df4],axis=1)
#data.to_excel('合并.xlsx')
```

```python
for i in range(data.shape[0]):
    if u'万' in data.薪资[i] and u'年' in data.薪资[i]:
        data.工资下限[i] = float(data.工资下限[i]) / 12 * 10
        data.工资上限[i] = float(data.工资上限[i]) / 12 * 10
    elif u'万' in data.薪资[i] and u'月' in data.薪资[i]:
        data.工资下限[i] = float(data.工资下限[i]) * 10
        data.工资上限[i] = float(data.工资上限[i]) * 10
```

```python
def get_salary(salary):
    if '-'in salary:  # 针对1-2万/月或者10-20万/年的情况，包含-
        low_salary = re.findall(re.compile('(\d*\.?\d+)'), salary)[0]
        high_salary = re.findall(re.compile('(\d?\.?\d+)'), salary)[1]
        if u'万' in salary and u'年' in salary:  # 单位统一成千/月的形式
            low_salary = float(low_salary) / 12 * 10
            high_salary = float(high_salary) / 12 * 10
        elif u'万' in salary and u'月' in salary:
            low_salary = float(low_salary) * 10
            high_salary = float(high_salary) * 10
    else:  # 针对20万以上/年和100元/天这种情况，不包含-，取最低工资，没有最高工资
        low_salary = re.findall(re.compile('(\d*\.?\d+)'), salary)[0]
        high_salary = ""
        if u'万' in salary and u'年' in salary:  # 单位统一成千/月的形式
            low_salary = float(low_salary) / 12 * 10
        elif u'万' in salary and u'月' in salary:
            low_salary = float(low_salary) * 10
        elif u'元'in salary and u'天'in salary:
            low_salary = float(low_salary)/1000*21  # 每月工作日21天
    return low_salary, high_salary
```
